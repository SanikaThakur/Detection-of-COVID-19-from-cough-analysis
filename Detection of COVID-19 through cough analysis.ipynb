```python
def plot_confusion_matrix(cm, classes, normalize=False,
                         title='Confusion matrix',
                         cmap=plt.cm.Blues):
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')
        
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i,j],
                horizontalalignment="center",
                color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

## Training Model

```python
X = data.drop(columns=['covid_status'])
```

```python
y = data['covid_status']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
NB_clf=MultinomialNB()
```

```python
NB_clf.fit(X_train, y_train)
pred = NB_clf.predict(X_test)
NB_score = metrics.accuracy_score(y_test, pred)
print("accuracy: %0.3f" % NB_score)
cm = metrics.confusion_matrix(y_test, pred)
plot_confusion_matrix(cm, classes=['negative', 'positive','positive_asymptomatic'])
```

```python
from sklearn.ensemble import RandomForestClassifier
RF_clf = RandomForestClassifier(n_estimators=100, 
                               bootstrap = True,
                               max_features = 'sqrt')
```

```python
RF_clf.fit(X_train, y_train)
pred = RF_clf.predict(X_test)
RF_score = metrics.accuracy_score(y_test, pred)
print("accuracy: %0.3f" % RF_score)
cm = metrics.confusion_matrix(y_test, pred)
plot_confusion_matrix(cm, classes=['negative', 'positive','positive_asymptomatic'])
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN_clf = KNeighborsClassifier(n_neighbors=5)
```

```python
KNN_clf.fit(X_train, y_train)
pred = KNN_clf.predict(X_test)
KNN_score = metrics.accuracy_score(y_test, pred)
print("accuracy: %0.3f" % KNN_score)
cm = metrics.confusion_matrix(y_test, pred)
plot_confusion_matrix(cm, classes=['negative', 'positive','positive_asymptomatic'])
```

```python
from sklearn.tree import DecisionTreeClassifier
DT_clf = DecisionTreeClassifier(criterion = "gini", 
            random_state = 100,max_depth=3, min_samples_leaf=5) 
```

```python
DT_clf.fit(X_train, y_train)
pred = DT_clf.predict(X_test)
DT_score = metrics.accuracy_score(y_test, pred)
print("accuracy: %0.3f" % DT_score)
cm = metrics.confusion_matrix(y_test, pred)
plot_confusion_matrix(cm, classes=['negative','positive','positive_asymptomatic'])
```
